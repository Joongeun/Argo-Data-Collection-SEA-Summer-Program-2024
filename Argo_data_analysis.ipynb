```python
import pandas as pd

#Extract only data points that are within the boundaries of each area
def limitSize(coors, df):
  avgs = []
  #min lon/lat, max lon/lat
  lon1, lat1, lon2, lat2 = coors
  df2 = pd.DataFrame(columns=df.columns)
  for ind in range(len(df)):
    if df["LONGITUDE"][ind] >= lon1 and df["LONGITUDE"][ind] <= lon2:
      if df["LATITUDE"][ind] >= lat1 and df["LATITUDE"][ind] <= lat2:
        df2.loc[len(df2.index)] = list(df.loc[ind])
  return df2

#Find average temp, salinity, and pressure
def findAvg(df):
  a = ["TEMP", "PSAL", "PRES"]
  l = []
  for i in a:
    l.append(df[i].mean())
  return l

#Save the avg. temp, salinity, pressure for years 2010-2020 for each area as a csv file
def saveCSV(l, ind):
  df = pd.DataFrame({"year": l["year"], k: l[k]})
  df.to_csv("avgs_for_area"+str(ind)+".csv")

#Areas 1-5 [min lon, min lat, max lon, max lat]
a = [[-180, -90, -72.61174335, 90], [-72.61174335, 39.9537, -65.29757095, 41.6362], [-72.61174335, -90, -65.29757095, 39.9537], [-65.29757095, 39.9537, 180, 41.6362], [-65.29757095, -90, 180, 39.9537]]
for ind, i in enumerate(a):
  avg = {"temp": [], "salinity": [], "pressure": [], "year": [i for i in range(2010, 2021)]}
  for year in range(2010, 2021):
    df = pd.read_csv("/content/"+str(year)+"_argo.csv")[["TEMP", "PSAL", "PRES", "LONGITUDE", "LATITUDE"]]
    #rectangular areas 1-5 (min lon/lat, max lon/lat) to find avg. temp, salinity, pressure
    df2 = limitSize(i, df)
    t, s, p = findAvg(df2)
    avg["temp"].append(t)
    avg["salinity"].append(s)
    avg["pressure"].append(p)
  print("Area:", ind+1, "done")
  df = pd.DataFrame(avg)
  df.to_csv("avgs_for_area"+str(ind+1)+".csv")

# df[::100].plot(x="TIME", y="TEMP", figsize=(18, 10))
# df.plot(x="TIME", y="TEMP", style = "bx", figsize=(12, 12))
```
